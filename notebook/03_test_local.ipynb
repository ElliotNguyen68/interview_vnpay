```python
data_rating=pd.read_csv('/Users/lap14171/interview/data/data_rating_vnpay.csv')
```

```python
data_rating
```

```python
customer_item_matrix=pd.pivot(data_rating,index='user_id',columns='item_id',values='score',)\
    .applymap(lambda x: 1 if x > 0 else 0)
```

```python
customer_item_matrix
```

```python
user_user_sim_matrix = pd.DataFrame(
    cosine_similarity(customer_item_matrix)
)
```

```python
user_user_sim_matrix.columns=customer_item_matrix.index
user_user_sim_matrix.index=customer_item_matrix.index
```

```python
user_user_sim_matrix
```

```python
user_user_sim_matrix.loc[[259308,259392,384027]]
```

```python
user_user_sim_matrix.loc[383922].sort_values(ascending=False)[1:4].values
```

```python
sim_users=user_user_sim_matrix.loc[383922].sort_values(ascending=False)
top_k_user_id=sim_users[1:4].index
top_k_sim_score=sim_users[1:4].values
```

```python
top_k_sim_score
```

```python
rec_items=set(customer_item_matrix.loc[lambda x:x.index.isin(top_k_user_id)].mul(top_k_sim_score,axis='index').sum(axis=0).sort_values(ascending=False)[:5].index.tolist())
```

```python
rec_items=customer_item_matrix.loc[lambda x:x.index.isin(top_k_user_id)].mul(top_k_sim_score,axis='index').sum(axis=0).sort_values(ascending=False)[:5].index.tolist()
```
